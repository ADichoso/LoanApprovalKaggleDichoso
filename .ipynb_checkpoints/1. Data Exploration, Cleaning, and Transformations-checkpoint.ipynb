#### 2.1.3.2. Feature Normalization

Another method for scaling the data is normalization, where we scale the range of values in a dataset to only within a certain range.
There will be two methods of normalization that will be explored in this dataset.

The first method will be scaling the values in a feature to only inhibit the range [0,1].

```python
def normalize(series):
        return (series - series.min()) / (series.max() - series.min())
```

```python
def normalize_test(series_train, series_test):
    return (series_test - series_train.min()) / (series_train.max() - series_train.min())
```

The second method is called standardization, which will make it so that a feature would have a mean of 0 and a variance of 1.

```python
def standardize(series):
    return (series - series.mean()) / series.std();
```

```python
def standardize_test(series_train, series_test):
    return (series_test - series_train.mean()) / series_train.std();
```

Additionally, it can be seen that this operation differs according to the train and test sets. I am scaling the values in the test set according to the ranges found in the train set. To perform this, we use the <a href="https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html">StandardScaler fit_transform() and transform()</a> methods, where the test data is fitted according to the mean and variance of the training data as to keep the scale consistent across both sets. The test set is not scaled according to its own parameters as it may lead to overfitting by <a href="https://www.geeksforgeeks.org/what-is-the-difference-between-transform-and-fit_transform-in-sklearn-python/">introducing bias into our model</a>. 

In my case, I wish to standardize all of the features. This is because normalization restricts the range of values found in a feature. This causes most of the data to be highly compressed due to the extreme values found in the features, of which there are only a few. I want my models to be able to effectively generalize the trends found in the data, and as such, I want to use standardization so as not to limit the range.

```python
cleaned_loans_test_df["person_emp_length"] = standardize_test(cleaned_loans_train_df["person_emp_length"], cleaned_loans_test_df["person_emp_length"])
cleaned_loans_test_df["loan_int_rate"] = standardize_test(cleaned_loans_train_df["loan_int_rate"], cleaned_loans_test_df["loan_int_rate"])
cleaned_loans_test_df["cb_person_cred_hist_length"] = standardize_test(cleaned_loans_train_df["cb_person_cred_hist_length"], cleaned_loans_test_df["cb_person_cred_hist_length"])
cleaned_loans_test_df["person_age"] = standardize_test(cleaned_loans_train_df["person_age"], cleaned_loans_test_df["person_age"])
cleaned_loans_test_df["person_income"] = standardize_test(cleaned_loans_train_df["person_income"], cleaned_loans_test_df["person_income"])
cleaned_loans_test_df["loan_amnt"] = standardize_test(cleaned_loans_train_df["loan_amnt"], cleaned_loans_test_df["loan_amnt"])
cleaned_loans_test_df["loan_percent_income"] = standardize_test(cleaned_loans_train_df["loan_percent_income"], cleaned_loans_test_df["loan_percent_income"])
```

## 3. Oversampling

Lastly, checking the number of samples in each class show that the distribution of class labels is imbalanced.

```python
# Count the frequency of each category
category_counts = loans_train_df['loan_status'].value_counts().sort_index()  # Sort alphabetically

# Plot the bar graph
plt.bar(category_counts.index, category_counts.values)

# Add labels and title
plt.xlabel('Loan Status')
plt.ylabel('Frequency')
plt.title('Loan status frequencies of Training Set entries')
plt.xticks([0, 1])
plt.savefig('labels_bargraph_train.png', bbox_inches='tight')
plt.show()
```

```python
cleaned_loans_train_df["loan_status"].value_counts()
```

Having an imbalance in the class labels may lead to the model lacking training in the minority class and skewing the classification results towards the majority class. As such, an oversampling technique may be implemented to balance the number of samples in each class.

From the oversampling methods discussed in our class, I wanted to explore the ADASYN method. I chose this method because the oversampling technique generates synthetic samples that are harder to classify and whose features are closer to the ones that majority classes have. this makes it so that in training, the distinction between the two classes may be identified more distinctly and may improve performance compared to other methods <a href="https://sci2s.ugr.es/keel/pdf/algorithm/congreso/2008-He-ieee.pdf">(He, et al., 2008)</a>.

```python
X = cleaned_loans_train_df.loc[:, cleaned_loans_train_df.columns != "loan_status"]
X = X.loc[:, X.columns != "id"]

y = cleaned_loans_train_df["loan_status"]

X.shape, y.shape
```

```python
from collections import Counter
from sklearn.datasets import make_classification
from imblearn.over_sampling import ADASYN

ada = ADASYN(sampling_strategy='minority')
X_res, y_res = ada.fit_resample(X, y)
X_res.shape, y_res.shape
```

```python
##Rebuild the train dataset
cleaned_loans_train_ada_df = pd.concat([X_res, y_res], axis=1)
cleaned_loans_train_ada_df["loan_status"].value_counts()
```
