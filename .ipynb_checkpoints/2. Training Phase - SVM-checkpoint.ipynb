```python
##Python libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
##Import Training Dataset
loans_train_df = pd.read_csv('./outputs/cleaned_loans_train.csv')
loans_train_df
```

```python
X = loans_train_df.loc[:, loans_train_df.columns != "loan_status"]
y = loans_train_df["loan_status"]
```

## Hyperparameter Tuning

```python
df_hyper_tuning = pd.DataFrame(columns=['C', 'kernel', 'gamma', 'tol', 'roc_auc'])
df_hyper_tuning
```

```python
from sklearn.model_selection import cross_val_score
from skopt import gp_minimize
from skopt.space import Real, Integer, Categorical
from skopt.utils import use_named_args
import numpy as np
import sys
# Define the search space
search_space = [
    Real(0.001, 3.0, name='C'),
    Categorical(['linear', 'poly', 'rbf', 'sigmoid', 'precomputed'], name='kernel'),
    Real(0.001, 1.0, name='gamma'),
    Real(0.0001, 0.1, name='tol'),
]

# Define your objective function (e.g., maximizing accuracy)
@use_named_args(search_space)
def objective_function(C, kernel, gamma, tol):
    print("================")
    print("Configuration:")
    print("C:", C)
    print("Kernel:", kernel)
    print("Gamma:", gamma)
    print("Tolerance:", tol)
    print("----------------")
    try:
        model = SVC(C=C, kernel=kernel, gamma=gamma, tol=tol, probability=True, verbose=True)
        roc_auc = cross_val_score(model, X, y, cv=3, scoring='roc_auc').mean()

        print("Results:", -roc_auc)
        df_hyper_tuning.loc[len(df_hyper_tuning.index)] = [C, kernel, gamma, tol, roc_auc] 
        print("================")
        return -roc_auc
    except:
        print("Invalid Config")
        return 100000
        

# Perform Bayesian Optimization
res = gp_minimize(objective_function, search_space, n_calls=50)

# Print best parameters
print("Best parameters:", res.x)
```

```python
df_hyper_tuning.to_csv('hyper_tuning/svm_hyper_tuning.csv', index=False, header=True, encoding='utf-8')
```

```python
clf = SVC(C=res.x[0], kernel=res.x[1], gamma=res.x[2], tol=res.x[3], probability=True)
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import RocCurveDisplay, roc_auc_score
import matplotlib.pyplot as plt

clf.fit(X,y)
# Perform cross-validation and predict probabilities for each fold
y_proba = cross_val_predict(clf, X, y, cv=5, method='predict_proba')[:, 1]

# Calculate the ROC AUC score
roc_auc = roc_auc_score(y, y_proba)

# Plot the ROC curve
RocCurveDisplay.from_predictions(y, y_proba)
plt.title(f"ROC Curve (AUC = {roc_auc:.2f})")
plt.show()
```

```python
from joblib import dump

dump(clf, './outputs/svm_model.joblib')
```
